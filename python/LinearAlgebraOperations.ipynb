```python
import numpy as np

#Dot product
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
dot = np.dot(a, b)
print("Dot product:", dot)
print("Alternative:", a @ b)
```

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
result = A @ B
print("\nMatrix multiply:")
print(result)
```

```python
# Determinant
det = np.linalg.det(A)
print("\nDeterminant:", det)
```

```python
# Inverse
inv = np.linalg.inv(A)
print("\nInverse:")
print(inv)
```

```python
# Verify: A @ inv = Identity
identity = A @ inv
print("\nA @ inv (should be I):")
print(identity.round(4))
```

```python
# Eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)
print("\nEigenvalues:", eigenvalues)
print("Eigenvectors")
print(eigenvectors)
```

```python
# Solve linear system: Ax = b
A = np.array([[2,1],[1,3]])
b = np.array([10, 15])
x = np.linalg.solve(A, b)
print("\nSolution to Ax=b:", x)
```

```python
# Vector norm (magnitude)
vector = np.array([3, 4])
magnitude = np.linalg.norm(vector)
print("\nVector magnitude", magnitude)
```

```python
# Create a 3x3 matrix representing lattice vectors. Calculate its determinant 
# (volume) and inverse. Verify the inverse is correct.

A = np.array([[1,2,3],[4,5,6],[7,8,1]])
det = np.linalg.det(A)
print("\nDeterminant:", det)

#inv
inv = np.linalg.inv(A)
print("\nInverse:", inv)

#verify
identity = A @ inv
print("\nA @ inv = \n", identity)

#check
print("\nIs it identity? ->", np.allclose(identity, np.eye(3)))
```
