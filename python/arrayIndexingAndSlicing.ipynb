```python
import numpy as np

#1D indexing
temps = np.array([900, 950, 1000, 1050, 1100])
print("First element:", temps[0])
print("Last element:", temps[-1])
print("First three:", temps[0:3])
print("Every other:", temps[::2])
```

```python
data = np.array ([
    [900, 30, 300],
    [1000, 50, 500],
    [1100, 70, 700],
    [1050, 55, 550]
])

print("\nElement [1,2]:", data[1,2])
print("\nFirst row:", data[0])
print("All diameters(col1):", data[:,1])
print("First two rows:", data[0:2])
```

```python
#Boolean indexing
high_temp = data[:,0]>1000
print("\nHigh temp experiments:")
print(data[high_temp])
```

```python
#Multiple conditions
good = (data[:, 0] > 950) & (data[:, 1] > 40)
print("\nGood conditions (T>950 & D>40):")
print(data[good])
```

```python
#Where function
large_idx = np.where(data[:, 1] > 50)
print("\nIndices where diameter > 50:", large_idx[0])
```

```python
#Modify arrays
modified = data.copy()
modified[modified[:, 1] < 50, 1] = 50
print("\nAfter setting min diameter to 50:")
print(modified)
```
