```python
import numpy as np

#Temperature convestion (vectorized!)
temps_c = np.array([900, 1000, 1100, 1200])
temps_k = temps_c + 273.15 #Converts All at once
print("Celcius:", temps_c)
print("Kelvin:", temps_k)
```

```python
#Element-wise operations
diameters = np.array ([30, 50, 70])
heights = np.array ([300, 500, 700])
aspect_ratios = heights / diameters
print("\nAspect ratios:", aspect_ratios)
```

```python
#Mathematical functions
radii = diameters / 2
volumes = np.pi * radii ** 2 * heights
print("Volumes (nm^3):", volumes.round(0))
```

```python
#Boolean operations
large_nanowires = diameters > 40
print("\nDiameter> 40nm:", large_nanowires)
```

```python
#Boolean indexing(filter)
large_diameters = diameters[diameters>40]
print("Large diameters only:", large_diameters)
```

```python
#Universal functions
temps = np.array([900, 1000, 1100])
exp_factor = np.exp(-1.5/(temps/1000))
print("\nExp factors:", exp_factor.round(4))
```

```python
np.random.seed(42)
random_diameters = np.random.uniform(40, 60, 10)
print("\nRandom diameters:", random_diameters.round(2))
random_heights = np.random.uniform(300, 700, 10)
print("\nRandom heights:", random_heights.round(2))

#convert diameters to radii
radii = random_diameters / 2

#calculate volumes
volumes = np.pi * radii ** 2 * random_heights

print("\nVolumes in (nm^3):", volumes.round(2))

#large volumes
large_volumes = volumes > 800000
print("\nLarge Volumes (nm^3):", large_volumes)
print("\nLarge Volume values(nm^3):", volumes[large_volumes].round(2))


print("\n===All Volumes===")
print("Volumes(nm^3):", volumes.round(0))
print("Mean volume:", volumes.mean().round(0), "nm^3")
print("Max volume:", volumes.max().round(0), "nm^3")
print("Min volume:", volumes.min().round(0), "nm^3")


```
