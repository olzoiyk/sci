```python
import numpy as np

#Reshape
data_1d = np.array([900, 30, 1000, 50, 1100, 70])
data_2d = data_1d.reshape(3,2)
print("Reshaped to 3x2:")
print(data_2d)
```

```python
flat = data_2d.flatten()
print("\nFlattened:", flat)
```

```python
#Transpose
print("\nTransposed:")
print(data_2d.T)
```

```python
# Stack arrays
temps = np.array([900, 1000, 1100])
diameters = np.array([30, 50, 70])
heights = np.array([300, 500, 700])

# Vertical stack(rows)
vstack = np.vstack([temps, diameters, heights])
print("\nVertical stack:")
print(vstack)
```

```python
# Column stack (easier for data)
cstack = np.column_stack([temps, diameters, heights])
print("\nColumn stack:")
print(cstack)
```

```python
# Concatenate
batch1 = np.array([[900, 30], [1000, 50]])
batch2 = np.array([[1100, 70], [1050, 55]])
combined = np.concatenate([batch1, batch2], axis=0)
print("\nConcatenated:")
print(combined)
```

```python
# Add dimension
temps_col = temps[:, np.newaxis]
print("\n1D shape:", temps.shape)
print("Column shape:", temps_col.shape)
```
